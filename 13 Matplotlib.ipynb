# 13. Matplotlib

## 13.1: Plotting a function

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return ((np.sin(x-2))**2)*(np.e**(-(x**2)))


xvals = np.arange(0, 2, 0.00001, dtype = float)
plt.plot(xvals, f(xvals), 'r-')
plt.title("sin^2(x-2)e^(-(x^2))")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
